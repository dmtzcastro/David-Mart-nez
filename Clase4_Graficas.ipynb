```python
df = pd.read_csv('Cartwheeldata.csv')
df.head()
```

```python
import seaborn as sns
```

```python
pip install seaborn
```

```python
sns.lmplot(x= 'Wingspan',
           y= 'CWDistance',
           data = df,
          )
plt.title('Scatter plot of wingspan vs cwdistance', fontsize = 15)
```

```python
sns.lmplot(x= 'Wingspan',
           y= 'CWDistance',
           data = df,
           fit_reg = False, #eliminar la linea de regression
           hue = 'Gender',  # para ver colores por cada categoria
          )
plt.title('Scatter plot of wingspan vs cwdistance', fontsize = 15)
```

```python
sns.swarmplot(x = 'Gender',
             y = 'CWDistance',
             data = df
            )
plt.title('Swarm plot of gender vs CWDistance',fontsize = 16)
plt.show()
```

## Con otra base de datos

### pregunta de investigacion
### la variable total bill tiene algun comportamiento de alguna distribucion?

```python
df = pd.read_csv('tips.csv')
df.head()
```

```python
sns.histplot(df['total_bill'],
             #kde=False #si se quita el kde = false, se pone la linea de tendencia
             #bins = 5 #especificar el numero de barras en el histograma
            ).set_title('histograma de total bill', fontsize = 16)
```

## conclusion:
### la varibale total_bills parece Weibull Distribution

```python
sns.boxplot(x = df['total_bill'],
           whis = (0,100) #sirve para quitar outliers
           ).set_title('boxplot de total bill',fontsize =17)
```

```python
sns.boxplot(x = df['total_bill'],
            y = df['sex'],
           whis = (0,100) #sirve para quitar outliers
           ).set_title('boxplot de total bill',fontsize =17)
plt.show
```
