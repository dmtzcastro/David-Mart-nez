```python
df = pd.read_csv('antropometria-dataset.csv')
df.head(2)
```

```python
pivottable = pd.pivot_table(df,
                            index = 'desc_ent',
                            values = 'peso',
                            aggfunc= 'mean'
                           )
                            
pivottable
```

```python
pivottable = pd.pivot_table(df,
                            index = ['desc_ent','sexo'] ,
                            values = 'peso',
                            aggfunc= 'mean'
                           )
                            
pivottable
```

```python
df1 =pd.read_csv('adult.csv')
df1.head(3)
```

```python
pt = pd.pivot_table(df1,
                    index = ['marital-status','education'],
                    values = 'hours-per-week',
                    aggfunc ='mean',
                   )
pt
```

```python
pt = pd.pivot_table(df1,
                    index = ['marital-status','sex','education'],
                    values = 'age',
                    aggfunc ='mean',
                   )
pt
```

```python
pt = pd.pivot_table(df1,
                    index = ['marital-status','sex'],
                    values = 'age',
                    columns = 'education',
                    aggfunc = 'median'
                   )
pt
```
