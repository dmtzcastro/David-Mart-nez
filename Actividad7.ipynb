```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
%ls
```

```python
df = pd.read_csv('antropometria-dataset-1.csv')
df.head(3)
```

```python
sns.lmplot(x= 'peso',
           y= 'cintura',
           data = df,
           fit_reg = False, #eliminar la linea de regression
           
          )
plt.title('peso vs cintura', fontsize = 15)
```

```python
sns.histplot(df['peso'],
             #kde=False #si se quita el kde = false, se pone la linea de tendencia
             #bins = 5 #especificar el numero de barras en el histograma
            ).set_title('histograma de peso', fontsize = 16)
```

```python
sns.histplot(df['cintura'],
             #kde=False #si se quita el kde = false, se pone la linea de tendencia
             #bins = 5 #especificar el numero de barras en el histograma
            ).set_title('histograma cintura', fontsize = 16)
```
