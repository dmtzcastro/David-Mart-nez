```python
df = pd.read_excel('Canada.xlsx',
                 sheet_name = 'Canada by Citizenship',
                 skiprows = range(20),
                 skipfooter = 2,
                 engine = 'openpyxl')
df.head(2)
```

```python
df.columns
```

```python
df.drop(['Type', 'Coverage', 'AREA','REG','DEV','DevName'],
          axis=1 , inplace = True)
```

```python
df.columns
```

```python
df.rename(columns = { 'OdName':'Country','AreaName':'Continent','RegName':'Region'},
          inplace = True)
df.columns
```

```python
all(isinstance(x,str) for x in df.columns)
```

```python
df.columns = list(map(str,df.columns))
df.columns
```

```python
df.iloc[:2,  3:]
```

```python
df['Total']=df.iloc[:, 3:].sum(axis = 1)
df.head(2)
```
