```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
%ls
```

```python
df = pd.read_csv('antropometria-dataset.csv')
df.head(2)
```

```python
sns.boxplot(data = df.loc[:,['peso']])
plt.xticks(fontsize = 17, rotation = 45)
plt.yticks(fontsize = 17)
plt.title('Boxplot',fontsize =17)
```

```python
sns.lmplot(x= 'peso',
           y= 'cintura',
           data = df,
           fit_reg = False, #eliminar la linea de regression
           hue = 'sexo',  # para ver colores por cada categoria
          )
plt.title('Scatter plot of peso vs cintura', fontsize = 15)
```

```python
sns.histplot(df['peso'],
            ).set_title('histograma de peso', fontsize = 16)
```

```python
sns.boxplot(data = df.loc[:,['talla']])
plt.xticks(fontsize = 17)
plt.yticks(fontsize = 17)
plt.title('Boxplot',fontsize =17)
```

```python
pivottable = pd.pivot_table(df,
                            index = 'desc_ent',
                            values= 'edad',
                            columns = 'sexo',
                            aggfunc= 'median'
                           )
                            
pivottable
```

```python
sns.histplot(df['cintura'],
             bins = 25
            ).set_title('histograma de peso', fontsize = 16)
```

```python
pivottable = pd.pivot_table(df,
                            index = 'desc_ent',
                            values= 'sistol',
                            columns = 'sexo',
                           )
                            
pivottable
```

```python
pivottable = pd.pivot_table(df,
                            index = 'desc_ent',
                            values= 'peso',
                            columns = 'sexo',
                            aggfunc = 'sum'
                           )
                            
pivottable
```
