```python
df = pd.read_csv('covid_north_america.csv')
df.head(2)
```

```python
pivot = pd.pivot_table(df,
                          index = 'Country/Other',
                          values = ['Total Deaths','Total Recovered'],
                          aggfunc ='mean'
                      )
pivot
```

```python
sns.boxplot(data = df.loc[:,'Total Deaths'])
plt.xticks(fontsize = 17, rotation = 45)
plt.yticks(fontsize = 17)
plt.title('Boxplots of certain variables',fontsize =17)
```
