```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%ls
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head(2)
```

```python
pivot = pd.pivot_table(df,
                          index = 'parental level of education',
                          columns = 'gender',
                          values = ['math score','reading score','writing score'],
                          aggfunc ='mean'
                      )
pivot
```

```python
pivot = pd.pivot_table(df,
                          index = 'race/ethnicity',
                       columns= 'test preparation course',
                          values = ['math score','reading score','writing score'],
                          aggfunc ='mean'
                      )
pivot
```

```python
sns.boxplot(data = df.loc[:,'math score'])
plt.xticks(fontsize = 17, rotation = 45)
plt.yticks(fontsize = 17)
plt.title('Boxplot Math Score',fontsize =17)
```

```python
sns.boxplot(x = df['math score'],
            y = df['gender'],
           whis = (0,100) #sirve para quitar outliers
           ).set_title('boxplot de total bill',fontsize =17)
plt.show
```

```python
sns.histplot(df['math score'],
             #kde=False #si se quita el kde = false, se pone la linea de tendencia
             #bins = 5 #especificar el numero de barras en el histograma
            ).set_title('histograma math score', fontsize = 16)
```
