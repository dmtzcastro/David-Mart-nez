```python
df = pd.read_excel('mic.xlsx',
                   engine = 'openpyxl'
                  )
df.head()
```

```python
counts = df['Age'].value_counts()
type(counts)
counts = counts.to_frame()
type(counts)

counts['%'] = (counts.iloc[:,:]/counts.loc[:,'Age'].sum(axis=0))*100
counts
```

```python
counts
```

```python
counts['Age'].plot(kind = 'pie')
```

```python
explode_list=[0.06,0.04,0.02,0.04]
color_list = ['gold','yellowgreen','coral','skyblue']
counts['Age'].plot(kind = 'pie',
                   figsize = (14,6),
                   startangle=90,
                   shadow = True,
                   labels = None,
                   autopct = '%1.1f%%',
                   pctdistance  =1.1,
                   explode = explode_list,
                  )
plt.legend(labels = counts.Age,
           bbox_to_anchor = (0.95,1.0)
          )
plt.title('Porcentajes grupos de edad',fontsize =17)
plt.show
```

```python
x = list(counts['Age_group'])
y = list(counts['Age'])
plt.bar(x,y,color='darkcyan')
plt.title('Grupos de edad',fontsize=17,y=1)

plt.show()
```

```python
age_genero = df.groupby(['Age','Sex']).size()
age_genero
age_genero = age_genero.to_frame()
age_genero
```

```python
age_genero_pivot=pd.pivot_table(index=['Age','Sex'], 
                                aggfunc = len,
                                data=df, values ='Availability')
                                      
age_genero_pivot             
```

```python
age_gender = age_genero_pivot.rename(columns = {'Availability':'Count'}
                                    ) 
age_gender['%'] = (age_gender.iloc[:,0]/age_gender.iloc[:,0].sum(axis=0))*100
age_gender
```

```python
sns.catplot(x='Age',
            data = df,
            kind = 'count',
            hue = 'Sex'
           )
```

```python
fig = px.sunburst(age_gender, 
            values = 'Count'
           )
fig.show()
#pendiente sig. clase
```
